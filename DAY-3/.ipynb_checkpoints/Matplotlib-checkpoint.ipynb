```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd 
```

```python
x = np.array([0, 6])
y = np.array([0, 250])

plt.plot(x,y)
plt.show()
```

```python
plt.plot(y, marker = 'o')
plt.show()
```

```python
plt.plot(y, 'o:r')
plt.show()
```

```python
plt.plot(y, marker = 'o', ms = 20, mfc = 'r')
plt.show()
```
